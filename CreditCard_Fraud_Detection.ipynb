# Data Preparation

The scaling is the process of maping values in a well defined interval. We use MinMaxScaler which use minimum and maxmim value of each feature and maps accordinlgy other values

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data['Amount'] = scaler.fit_transform(data['Amount'].values.reshape(-1,1))
data
```

The data may contain missing or undefined values. This may cause training erro while training the model. So by using dataframe dropna function we are removing all these transactions.

# Modeling

<b>Data Grouping<b>

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,:-1], data.iloc[:,-1], test_size=0.33, random_state=42)
```

```python
X_train
```

<b>Models<b>

```python
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import matthews_corrcoef
from matplotlib import pyplot as plt
```

<b>Linear Discriminant Analysis</b>

<b>Logistic Regression</b> 

```python
%%time
from sklearn.linear_model import LogisticRegression

LG_model = LogisticRegression()
LG_model.fit(X_train, y_train)
LG_predict = LG_model.predict(X_test)
print('Logistic Regression Accuracy: {}'.format(LG_model.score(X_test, y_test)))

plt.figure()
plot_confusion_matrix(LG_model, X_test, y_test,
                                 display_labels=['Normal', 'Fraudulent'],
                                 cmap=plt.cm.Blues,
                                 normalize=None)
plt.title('Logistic Regression Confusion Metrix')
plt.show()
MCC=matthews_corrcoef(y_test, LG_predict)
print('Logistic Regression')
print(classification_report(y_test, LG_predict))
print('Matthews Correlation Coefficient: {}'.format(MCC))
```

<b>Support Vector Machine</b>

```python
%%time
from sklearn.svm import SVC

SVC_model = SVC()
SVC_model.fit(X_train, y_train)
print('Support Vector Machine Accuracy: {}'.format(SVC_model.score(X_test, y_test)))
SVC_predict = SVC_model.predict(X_test)

plt.figure()
plot_confusion_matrix(SVC_model, X_test, y_test,
                                 display_labels=['Normal', 'Fraudulent'],
                                 cmap=plt.cm.Blues,
                                 normalize=None)
plt.title('Support Vector Machine Confusion Metrix')
plt.show()
MCC=matthews_corrcoef(y_test, SVC_predict)
print('Support Vector Machine')
print(classification_report(y_test, SVC_predict))
print('Matthews Correlation Coefficient: {}'.format(MCC))
```

<b>Naive Bayes (Gaussian)</b>

```python
%%time
from sklearn.naive_bayes import GaussianNB

NB_model = GaussianNB()
NB_model.fit(X_train, y_train)
print('Naive Bayes Accuracy: {}'.format(NB_model.score(X_test, y_test)))
NB_predict = NB_model.predict(X_test)

plt.figure()
plot_confusion_matrix(NB_model, X_test, y_test,
                                 display_labels=['Normal', 'Fraudulent'],
                                 cmap=plt.cm.Blues,
                                 normalize=None)
plt.title('Naive Bayes Confusion Metrix')
plt.show()
MCC=matthews_corrcoef(y_test, NB_predict)
print('Naive Bayes')
print(classification_report(y_test, NB_predict))
print('Matthews Correlation Coefficient: {}'.format(MCC))
```

<b>K-Nearest Neighbors</b>

```python
%%time
from sklearn.neighbors import KNeighborsClassifier

KNN_model = KNeighborsClassifier()
KNN_model.fit(X_train, y_train)
print('K-Nearest Neighbors Accuracy: {}'.format(KNN_model.score(X_test, y_test)))
KNN_predict = KNN_model.predict(X_test)

plt.figure()
plot_confusion_matrix(KNN_model, X_test, y_test,
                                 display_labels=['Normal', 'Fraudulent'],
                                 cmap=plt.cm.Blues,
                                 normalize=None)
plt.title('K-Nearest Neighbors Confusion Metrix')
plt.show()
MCC=matthews_corrcoef(y_test, KNN_predict)
print('K-Nearest Neighbors')
print(classification_report(y_test, KNN_predict))
print('Matthews Correlation Coefficient: {}.'.format(MCC))
```

<b>Decision Tree<b>

```python
%%time
from sklearn.tree import DecisionTreeClassifier

DT_model = DecisionTreeClassifier()
DT_model.fit(X_train, y_train)
print('Decision Tree Accuracy: {}'.format(DT_model.score(X_test, y_test)))
DT_predict = DT_model.predict(X_test)

plt.figure()
plot_confusion_matrix(DT_model, X_test, y_test,
                                 display_labels=['Normal', 'Fraudulent'],
                                 cmap=plt.cm.Blues,
                                 normalize=None)
plt.title('Decision Tree Confusion Metrix')
plt.show()
MCC=matthews_corrcoef(y_test, DT_predict)
print('Decision Tree')
print(classification_report(y_test, DT_predict))
print('Matthews Correlation Coefficient: {}.'.format(MCC))
```

<b>Extreem Gradient Boost<b>

```python
%%time
from xgboost import XGBClassifier

XGB_model = XGBClassifier()
XGB_model.fit(X_train, y_train)
print('Extreem Gradient Boost Accuracy: {}'.format(XGB_model.score(X_test, y_test)))
XGB_predict = XGB_model.predict(X_test)

plt.figure()
plot_confusion_matrix(XGB_model, X_test, y_test,
                                 display_labels=['Normal', 'Fraudulent'],
                                 cmap=plt.cm.Blues,
                                 normalize=None)
plt.title('Extreem Gradient Boost Confusion Metrix')
plt.show()
MCC=matthews_corrcoef(y_test, XGB_predict)
print('Extreem Grabient Boost')
print(classification_report(y_test, XGB_predict))
print('Matthews Correlation Coefficient: {}.'.format(MCC))
```

<b>Random Forest<b>

```python
%%time
from sklearn.ensemble import RandomForestClassifier

RF_model = RandomForestClassifier()
RF_model.fit(X_train, y_train)
print('Random Forest Classifier Accuracy: {}'.format(RF_model.score(X_test, y_test)))
RF_predict = RF_model.predict(X_test)

plt.figure()
plot_confusion_matrix(RF_model, X_test, y_test,
                                 display_labels=['Normal', 'Fraudulent'],
                                 cmap=plt.cm.Blues,
                                 normalize=None)
plt.title('Random Forest Classifier Confusion Metrix')
plt.show()
MCC=matthews_corrcoef(y_test, RF_predict)
print('Random Forest Classifier')
print(classification_report(y_test, RF_predict))
print('Matthews Correlation Coefficient: {}.'.format(MCC))
```
